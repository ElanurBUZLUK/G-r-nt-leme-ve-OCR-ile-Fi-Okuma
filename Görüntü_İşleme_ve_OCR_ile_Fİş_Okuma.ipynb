## 5. Görüntü Ön İşleme (OpenCV) — Rotasyon + Çoklu Strateji

Gerçek fiş fotoğraflarında en sık görülen problem, görüntünün **90/180/270 derece dönmüş** olması veya preprocessing'in yazıyı bozmasıdır. Bu durumda OCR motoru anlamlı kelimeler yerine `8`, `{`, `;`, `ğ` gibi tek karakterler döndürebilir.

Bu notebook bu problemi açıkça formüle eder:

```text
4 rotasyon × 4 preprocessing stratejisi = 16 OCR adayı
```

Her aday OCR'a verilir. Sonra aday metinler şu sinyallerle skorlanır:

- Alfanümerik karakter sayısı
- Anlamlı market/fiş kelimesi sayısı: `TOPLAM`, `KDV`, `TARIH`, `FIS`, `BIM`, `A101`, `MIGROS`, `SOK`
- Para formatı sayısı: `24,90`, `245.90`
- Tarih/saat formatı sayısı
- Tek karakter gürültü oranı

En yüksek skorlu aday final OCR metni olarak seçilir.


````python
def load_image_safe(image_path):
    """
    Görseli OpenCV ile okur.
    Okunamazsa None döner; pipeline çökmek yerine hata kaydı üretir.
    """
    bgr = cv2.imread(str(image_path))
    if bgr is None:
        return None, None
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    return bgr, rgb


def rotate_bgr(bgr, angle):
    """
    WhatsApp / telefon fotoğraflarında EXIF veya manuel rotasyon problemi olabilir.
    Bu yüzden 0, 90, 180, 270 derece adayları açıkça denenir.
    """
    if angle == 0:
        return bgr.copy()
    if angle == 90:
        return cv2.rotate(bgr, cv2.ROTATE_90_CLOCKWISE)
    if angle == 180:
        return cv2.rotate(bgr, cv2.ROTATE_180)
    if angle == 270:
        return cv2.rotate(bgr, cv2.ROTATE_90_COUNTERCLOCKWISE)
    raise ValueError(f'Desteklenmeyen rotasyon açısı: {angle}')


def deskew(gray):
    """
    Hafif eğik fişleri düzeltir.
    90 derece rotasyonu çözmez; o iş rotate_bgr() ile 4 yön denenerek yapılır.
    """
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)
    lines = cv2.HoughLines(edges, 1, np.pi / 180, threshold=80)
    if lines is None:
        return gray

    angles = []
    for rho, theta in lines[:, 0]:
        angle = np.degrees(theta) - 90
        if abs(angle) < 15:  # küçük eğiklikleri düzelt, büyük dönüşleri rotasyon adaylarına bırak
            angles.append(angle)

    if not angles:
        return gray

    median_angle = float(np.median(angles))
    if abs(median_angle) < 0.3:
        return gray

    h, w = gray.shape
    M = cv2.getRotationMatrix2D((w // 2, h // 2), median_angle, 1.0)
    return cv2.warpAffine(gray, M, (w, h), flags=cv2.INTER_LINEAR,
                          borderMode=cv2.BORDER_REPLICATE)


def make_strategy_clahe(gray, scale=1.6):
    """Resize → CLAHE → hafif denoise → deskew → adaptive threshold."""
    g = cv2.resize(gray, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    g = clahe.apply(g)
    g = cv2.fastNlMeansDenoising(g, h=7, templateWindowSize=7, searchWindowSize=21)
    g = deskew(g)
    g = cv2.adaptiveThreshold(g, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                               cv2.THRESH_BINARY, 31, 9)
    g = cv2.morphologyEx(g, cv2.MORPH_CLOSE, np.ones((2, 1), np.uint8))
    return g


def make_strategy_otsu(gray, scale=1.6):
    """Resize → CLAHE → blur → deskew → Otsu threshold."""
    g = cv2.resize(gray, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    g = clahe.apply(g)
    g = cv2.GaussianBlur(g, (3, 3), 0)
    g = deskew(g)
    _, g = cv2.threshold(g, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return g


def make_strategy_minimal(gray, scale=1.6):
    """Threshold fazla bozuyorsa daha az müdahaleli fallback."""
    g = cv2.resize(gray, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    g = deskew(g)
    g = cv2.adaptiveThreshold(g, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                               cv2.THRESH_BINARY, 21, 5)
    return g


def make_strategy_direct(gray, scale=1.6):
    """Yalnızca resize + CLAHE. EasyOCR bazen threshold'suz gri görüntüde daha iyi çalışır."""
    g = cv2.resize(gray, None, fx=scale, fy=scale, interpolation=cv2.INTER_CUBIC)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    g = clahe.apply(g)
    g = deskew(g)
    return g


def preprocess_variants_for_rotation(bgr_rotated):
    """Tek rotasyon için 4 preprocessing varyantı üretir."""
    gray = cv2.cvtColor(bgr_rotated, cv2.COLOR_BGR2GRAY)
    return {
        'clahe_adaptive': make_strategy_clahe(gray),
        'clahe_otsu':     make_strategy_otsu(gray),
        'minimal':        make_strategy_minimal(gray),
        'direct_clahe':   make_strategy_direct(gray),
    }


def preprocess_all_strategies(image_path):
    """
    4 rotasyon × 4 preprocessing stratejisi üretir.
    Dönüş: original_bgr, original_rgb, strategy_dict
    strategy_dict key örneği: rot90_clahe_adaptive
    """
    bgr, rgb = load_image_safe(image_path)
    if bgr is None:
        return None, None, {}

    strategies = {}
    for angle in [0, 90, 180, 270]:
        rbgr = rotate_bgr(bgr, angle)
        variants = preprocess_variants_for_rotation(rbgr)
        for variant_name, processed in variants.items():
            strategies[f'rot{angle}_{variant_name}'] = processed

    return bgr, rgb, strategies


def visualize_preprocessing(image_path, max_show=8):
    """
    Orijinal görseli ve üretilen OCR adaylarından örnekleri gösterir.
    16 adayın tamamını basmak notebook'u kalabalıklaştırabileceği için ilk 8 gösterilir.
    """
    bgr, rgb, strategies = preprocess_all_strategies(image_path)
    if bgr is None:
        print('Görsel yüklenemedi.')
        return

    items = list(strategies.items())[:max_show]
    fig, axes = plt.subplots(1, len(items) + 1, figsize=(4 * (len(items) + 1), 6))
    axes[0].imshow(rgb)
    axes[0].set_title('Orijinal')
    axes[0].axis('off')

    for ax, (name, img) in zip(axes[1:], items):
        ax.imshow(img, cmap='gray')
        ax.set_title(name, fontsize=8)
        ax.axis('off')

    plt.suptitle(f'Preprocessing/Rotasyon Adayları: {Path(image_path).name}', fontsize=12)
    plt.tight_layout()
    plt.show()


if image_paths:
    visualize_preprocessing(image_paths[0])
````

## 5.5 Metin Normalizasyonu — Erken Tanımlama

`full_normalize` fonksiyonu OCR kalite skorlaması (Bölüm 6) tarafından kullanılır.
Bu nedenle OCR hücresinden **önce** çalıştırılmalıdır.

Normalizasyon adımları:
- Büyük harfe çevirme
- Türkçe → ASCII karakter dönüşümü (`İ→I`, `Ş→S` …)
- OCR anahtar kelime hata düzeltmeleri (`T0PLAM→TOPLAM`, `K.D.V→KDV` …)


````python
# ── Metin Normalizasyon Fonksiyonları ─────────────────────────────────────────
# Bu fonksiyonlar OCR kalite skorlaması (Bölüm 6) tarafından kullanılır.
# Bu nedenle OCR hücresinden ÖNCE tanımlanmak zorundadır.

def normalize_tr(text):
    """
    Büyük harfe çevirir ve Türkçe karakterleri ASCII eşdeğerine dönüştürür.
    Regex kalıpları ASCII'de çalıştığı için bu adım zorunludur.
    """
    if not text:
        return ''
    text = str(text).upper()
    tr_map = {
        'İ': 'I', 'Ş': 'S', 'Ğ': 'G',
        'Ü': 'U', 'Ö': 'O', 'Ç': 'C',
        '\u00a0': ' '
    }
    for k, v in tr_map.items():
        text = text.replace(k, v)
    text = __import__('re').sub(r'[ \t]+', ' ', text)
    return text


def fix_ocr_keywords(text):
    """
    Bilinen OCR hata varyasyonlarını düzeltir.
    Yalnızca anahtar kelimeler üzerinde güvenli değişimler yapılır —
    sayısal alanlara dokunulmaz.
    """
    import re as _re
    text = _re.sub(r'\bT[O0]P[L1I][A4]M\.?\b', 'TOPLAM', text)
    text = _re.sub(r'\bTOPI_?AM\b', 'TOPLAM', text)
    text = _re.sub(r'\bT[O0]P\.\b', 'TOPLAM', text)
    text = _re.sub(r'\bGENEL\s+T[O0]PLAM\b', 'GENEL TOPLAM', text)
    text = _re.sub(r'\bK\.?D\.?V\.?\b', 'KDV', text)
    text = _re.sub(r'\bTOPKDV\b', 'TOPLAM KDV', text)
    text = _re.sub(r'\bKDV\s+TOPLAMI\b', 'TOPLAM KDV', text)
    text = _re.sub(r'\b[O0]DENECEK\b', 'ODENECEK', text)
    text = _re.sub(r'\bNAKIT\s+[O0]DEME\b', 'NAKIT', text)
    return text


def full_normalize(raw_text):
    """Pipeline normalizasyon adımlarını sırayla uygular."""
    t = normalize_tr(raw_text)
    t = fix_ocr_keywords(t)
    return t


print('✓ Normalizasyon fonksiyonları tanımlandı (full_normalize hazır).')
````

````python

def run_ocr_on_image(processed_img, min_confidence=OCR_MIN_CONFIDENCE):
    """
    Tek OCR aday görüntüsü üzerinde EasyOCR çalıştırır.

    Not:
    - Düşük confidence satırları tamamen kaybetmiyoruz.
    - Parsing için kullanılan raw_text'e sadece minimum eşiği geçen satırlar alınır.
    - Detay tablosunda bbox/confidence/low_confidence bilgisi saklanır.
    """
    results = reader.readtext(processed_img, detail=1, paragraph=False)

    rows = []
    used_texts = []

    for idx, (bbox, text, confidence) in enumerate(results):
        conf = float(confidence)
        text = str(text).strip()
        used = conf >= min_confidence and len(text) > 0

        rows.append({
            'line_id':        idx,
            'text':           text,
            'confidence':     round(conf, 4),
            'bbox':           bbox,
            'low_confidence': conf < OCR_LOW_CONF_FLAG,
            'used_in_text':   used,
        })

        if used:
            used_texts.append(text)

    raw_text = '\n'.join(used_texts)
    return raw_text, rows


QUALITY_KEYWORDS = [
    'TOPLAM', 'T0PLAM', 'TUTAR', 'GENEL', 'ARA', 'KDV',
    'FIS', 'FIŞ', 'BELGE', 'TARIH', 'SAAT',
    'BIM', 'BİM', 'A101', 'MIGROS', 'MİGROS', 'SOK', 'ŞOK',
    'VERGI', 'VKN', 'VD', 'NAKIT', 'KREDI', 'KART', 'PARA USTU',
    'KASIYER', 'KASİYER', 'SUBE', 'MAGAZA'
]

FINANCIAL_KEYWORDS = ['TOPLAM', 'KDV', 'TUTAR', 'NAKIT', 'KREDI', 'KART', 'PARA USTU']
MARKET_KEYWORDS = ['BIM', 'BİM', 'A101', 'MIGROS', 'MİGROS', 'SOK', 'ŞOK']


def _safe_mean(values, default=0.0):
    values = list(values)
    return float(np.mean(values)) if values else default


def _text_noise_features(raw_text, rows):
    """
    OCR çıktısının çöp/gürültü olma ihtimalini ölçen yardımcı metrikler.
    """
    used_rows = [r for r in rows if r.get('used_in_text')]
    texts = [str(r.get('text', '')).strip() for r in used_rows]
    tokens = re.findall(r'\S+', raw_text or '')

    one_char_lines = sum(1 for t in texts if len(t) <= 1)
    two_char_lines = sum(1 for t in texts if len(t) <= 2)
    alpha_tokens = re.findall(r'[A-ZÇĞIİÖŞÜa-zçğıöşü]{2,}', raw_text or '')
    numeric_tokens = re.findall(r'\d+', raw_text or '')
    weird_tokens = [t for t in tokens if not re.search(r'[A-Za-zÇĞİÖŞÜçğıöşü0-9]', t)]

    total_chars = max(len(raw_text or ''), 1)
    alnum_chars = len(re.findall(r'[A-Za-zÇĞİÖŞÜçğıöşü0-9]', raw_text or ''))
    alnum_ratio = alnum_chars / total_chars

    return {
        'used_line_count': len(used_rows),
        'token_count': len(tokens),
        'one_char_lines': one_char_lines,
        'two_char_lines': two_char_lines,
        'alpha_token_count': len(alpha_tokens),
        'numeric_token_count': len(numeric_tokens),
        'weird_token_count': len(weird_tokens),
        'alnum_ratio': alnum_ratio,
        'avg_confidence': _safe_mean([r.get('confidence', 0) for r in used_rows], 0.0),
    }


def product_like_line_count(raw_text):
    """
    Ürün satırına benzeyen satırları sayar.
    Ürün adı + para deseni içeren satırlar OCR kalitesi için güçlü sinyaldir.
    """
    count = 0
    for line in str(raw_text or '').split('\n'):
        n = full_normalize(line)
        has_letters = bool(re.search(r'[A-ZÇĞIİÖŞÜ]{2,}', n))
        has_money = bool(re.search(r'\d{1,7}(?:[.]\d{3})*[,.]\d{2}', n))
        blocked = any(k in n for k in ['TOPLAM', 'KDV', 'NAKIT', 'KREDI', 'PARA USTU', 'TARIH', 'SAAT'])
        if has_letters and has_money and not blocked:
            count += 1
    return count


def ocr_quality_score(raw_text, rows):
    """
    V4 OCR aday skoru.

    Amaç:
    - Yanlış rotasyon/preprocessing sonucunda oluşan tek karakterli gürültüyü cezalandırmak
    - Fişe özgü sinyalleri ödüllendirmek
    - Sadece line count'a bağlı kalmamak
    """
    if not raw_text or not str(raw_text).strip():
        return 0.0

    norm = full_normalize(raw_text)
    noise = _text_noise_features(raw_text, rows)

    money_count = len(re.findall(r'\d{1,7}(?:[.]\d{3})*[,.]\d{2}', norm))
    date_count = len(re.findall(r'\d{2}[./-]\d{2}[./-]\d{2,4}', norm))
    time_count = len(re.findall(r'\d{1,2}:\d{2}', norm))
    keyword_count = sum(1 for kw in QUALITY_KEYWORDS if kw in norm)
    financial_keyword_count = sum(1 for kw in FINANCIAL_KEYWORDS if kw in norm)
    market_keyword_count = sum(1 for kw in MARKET_KEYWORDS if kw in norm)
    product_line_count = product_like_line_count(raw_text)

    # Çöp OCR çıktısı: çok fazla tek karakter, anlamsız token, düşük alnum oranı.
    noise_penalty = (
        noise['one_char_lines'] * 12 +
        noise['two_char_lines'] * 3 +
        noise['weird_token_count'] * 4 +
        max(0, 0.55 - noise['alnum_ratio']) * 60
    )

    score = (
        noise['avg_confidence'] * 35 +
        min(noise['alpha_token_count'], 80) * 1.2 +
        min(noise['numeric_token_count'], 80) * 0.8 +
        money_count * 28 +
        date_count * 36 +
        time_count * 22 +
        keyword_count * 18 +
        financial_keyword_count * 18 +
        market_keyword_count * 20 +
        product_line_count * 16 +
        min(noise['used_line_count'], 60) * 0.5 -
        noise_penalty
    )

    return round(float(max(score, 0.0)), 3)


def run_ocr_best_strategy(bgr, strategies, return_all_candidates=False):
    """
    16 OCR adayını dener ve V4 kalite skoruna göre en iyi adayı seçer.
    """
    best = {
        'text': '',
        'rows': [],
        'strategy': 'none',
        'score': -1.0,
    }
    candidate_scores = []

    for strategy_name, processed in strategies.items():
        try:
            raw_text, rows = run_ocr_on_image(processed)
            score = ocr_quality_score(raw_text, rows)
            noise = _text_noise_features(raw_text, rows)

            candidate_scores.append({
                'strategy': strategy_name,
                'score': score,
                'line_count': len(rows),
                'used_line_count': noise['used_line_count'],
                'avg_confidence': round(noise['avg_confidence'], 4),
                'one_char_lines': noise['one_char_lines'],
                'money_count': len(re.findall(r'\d{1,7}(?:[.]\d{3})*[,.]\d{2}', full_normalize(raw_text))),
                'keyword_hits': sum(1 for kw in QUALITY_KEYWORDS if kw in full_normalize(raw_text)),
                'product_like_lines': product_like_line_count(raw_text),
                'text_preview': raw_text[:140].replace('\n', ' '),
            })

            if score > best['score']:
                best = {
                    'text': raw_text,
                    'rows': rows,
                    'strategy': strategy_name,
                    'score': score,
                }
        except Exception as e:
            candidate_scores.append({
                'strategy': strategy_name,
                'score': -1,
                'line_count': 0,
                'used_line_count': 0,
                'avg_confidence': 0,
                'one_char_lines': None,
                'money_count': 0,
                'keyword_hits': 0,
                'product_like_lines': 0,
                'text_preview': f'OCR hatası: {e}',
            })

    # Son fallback: preprocessing her şeyi bozduysa orijinal görüntüyü de dene.
    if bgr is not None:
        try:
            raw_text, rows = run_ocr_on_image(bgr)
            score = ocr_quality_score(raw_text, rows)
            noise = _text_noise_features(raw_text, rows)
            candidate_scores.append({
                'strategy': 'original_bgr',
                'score': score,
                'line_count': len(rows),
                'used_line_count': noise['used_line_count'],
                'avg_confidence': round(noise['avg_confidence'], 4),
                'one_char_lines': noise['one_char_lines'],
                'money_count': len(re.findall(r'\d{1,7}(?:[.]\d{3})*[,.]\d{2}', full_normalize(raw_text))),
                'keyword_hits': sum(1 for kw in QUALITY_KEYWORDS if kw in full_normalize(raw_text)),
                'product_like_lines': product_like_line_count(raw_text),
                'text_preview': raw_text[:140].replace('\n', ' '),
            })
            if score > best['score']:
                best = {
                    'text': raw_text,
                    'rows': rows,
                    'strategy': 'original_bgr',
                    'score': score,
                }
        except Exception:
            pass

    if return_all_candidates:
        return best['text'], best['rows'], best['strategy'], best['score'], candidate_scores

    return best['text'], best['rows'], best['strategy'], best['score']


if image_paths:
    original_bgr, _, sample_strategies = preprocess_all_strategies(image_paths[0])
    sample_text, sample_rows, chosen_strategy, chosen_score, sample_candidates = run_ocr_best_strategy(
        original_bgr, sample_strategies, return_all_candidates=True
    )

    print(f'Seçilen strateji : {chosen_strategy}')
    print(f'OCR kalite skoru : {chosen_score}')
    print(f'Tespit edilen satır: {len(sample_rows)}')
    print(f'Kullanılan satır : {sum(1 for r in sample_rows if r["used_in_text"])}')
    print('\nEn iyi 8 OCR adayı:')
    display(pd.DataFrame(sample_candidates).sort_values('score', ascending=False).head(8))
    print('\n─── Ham OCR Metin ───')
    print(sample_text[:1200])
````

````python
# ── OCR Bounding Box Görselleştirme ─────────────────────────────────────────
# Yeşil = yüksek confidence, Kırmızı = düşük confidence.
# "Hangi satırı nerede okudu?" sorusuna görsel yanıt verir.

def visualize_ocr_bboxes(bgr, rows, title='OCR Bounding Boxes'):
    """
    OCR tespit edilen alanları görsel üzerine çizer.
    Debugging ve sunumda nerede okunduğunu doğrulamak için kullanılır.
    """
    if bgr is None or not rows:
        print('Görselleştirilecek veri yok.')
        return

    vis = cv2.cvtColor(bgr.copy(), cv2.COLOR_BGR2RGB)

    for row in rows:
        if not row.get('used_in_text', False):
            continue
        bbox = np.array(row['bbox'], dtype=np.int32)
        color = (220, 50, 50) if row['low_confidence'] else (50, 200, 50)
        cv2.polylines(vis, [bbox], isClosed=True, color=color, thickness=2)
        x0, y0 = int(bbox[0][0]), int(bbox[0][1])
        label = f"{row['text'][:16]} [{row['confidence']:.2f}]"
        cv2.putText(vis, label, (x0, max(y0 - 4, 12)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.38, color, 1, cv2.LINE_AA)

    plt.figure(figsize=(10, 13))
    plt.imshow(vis)
    plt.title(title, fontsize=11)
    plt.axis('off')
    green = mpatches.Patch(color=(50/255, 200/255, 50/255), label='Yüksek confidence')
    red   = mpatches.Patch(color=(220/255, 50/255, 50/255), label='Düşük confidence')
    plt.legend(handles=[green, red], loc='upper right')
    plt.tight_layout()
    plt.show()


if image_paths:
    sample_path = image_paths[0]   # viz hücresi bu değişkeni kullanır

    visualize_ocr_bboxes(
        original_bgr,
        sample_rows,
        title=f'OCR BBoxes — {sample_path.name}'
    )

    # OCR satır detay tablosu
    df_sample_ocr = pd.DataFrame(sample_rows)
    print('\nOCR Satır Detayları (ilk 25):')
    display(
        df_sample_ocr[['line_id', 'text', 'confidence', 'low_confidence', 'used_in_text']]
        .head(25)
    )
````

## 8. Regex + Heuristic Parsing — V4 Geniş Fiş Şeması

Bu bölümde regex extraction yalnızca `Tarih / Toplam / KDV` ile sınırlı kalmaz. Fişte görülebilecek alanlar geniş bir şema altında yakalanır:

V4 farkı: parsing sonucunda ayrıca **field-level confidence** ve **consistency validation** üretilir.


### KDV çıkarımı güçlendirme notu

Bu sürümde KDV parser özellikle kabul kriterini yükseltmek için güncellendi:

- `%1`, `%8`, `%10`, `%18`, `%20` oranları desteklenir.
- `TOPLAM KDV`, `KDV TUTARI`, `TOPKDV` gibi doğrudan satırlar aranır.
- KDV tablosunda birden fazla tutar varsa son tutar KDV adayı kabul edilir.
- Nihai KDV seçimi, genel toplam ile tutarlılık kontrolünden geçirilir.


````python

AMT = r'(\d{1,7}(?:[.]\d{3})*[,.]\d{2}|\d{1,7}[,.]\d{2})'

DATE_PATTERNS = [
    r'\b(\d{2}[.]\d{2}[.]\d{4})\b',
    r'\b(\d{2}[/]\d{2}[/]\d{4})\b',
    r'\b(\d{2}[-]\d{2}[-]\d{4})\b',
    r'\b(\d{2}[.]\d{2}[.]\d{2})\b',
    r'\b(\d{4}[.-]\d{2}[.-]\d{2})\b',
]
TIME_PATTERNS = [
    r'\b(\d{1,2}:\d{2}(?::\d{2})?)\b',
]

MARKET_PATTERNS = [
    r'\b(BIM|B[I1]M|B\.I\.M)\b',
    r'\b(A101|A\s*101)\b',
    r'\b(MIGROS|M[I1]GROS|MİGROS)\b',
    r'\b(SOK|ŞOK|S[O0]K)\b',
]

TAX_PATTERNS = [
    r'\b(?:VERGI\s*NO|VERGI\s*NUMARASI|VKN|VD)\s*[:\-]?\s*(\d{8,11})\b',
    r'\b(\d{10})\b',
]

RECEIPT_NO_PATTERNS = [
    # Fiş/Belge No: mutlaka rakamla başlamalı — 'FATURA/TRSAL' gibi metinleri dışlar
    r'\b(?:FIS|FIŞ|BELGE)\s*(?:NO|NU|NUMARA|#)?\s*[:\-]?\s*(\d[A-Z0-9\-\/]{1,18})\b',
    r'\bFATURA\s+(?:NO|NU|NUMARA)\s*[:\-]?\s*(\d[A-Z0-9\-\/]{1,18})\b',
    r'\bNO\s*[:\-]?\s*(\d{4,})\b',
]

CASHIER_PATTERNS = [
    r'\b(?:KASIYER|KASIER|KASİYER)\s*[:\-]?\s*([A-ZÇĞIİÖŞÜ ]{2,30})\b',
    r'\b(?:KAS\.?)\s*[:\-]?\s*([A-ZÇĞIİÖŞÜ ]{2,30})\b',
]

SUBTOTAL_KEYWORDS = ['ARA TOPLAM', 'ARATOPLAM', 'TOPLAM TUTAR']
TOTAL_VAT_KEYWORDS = ['TOPLAM KDV', 'KDV TOPLAM', 'KDV TUTARI', 'TOPKDV']
GENERAL_TOTAL_KEYWORDS = ['GENEL TOPLAM', 'ODENECEK TUTAR', 'ODENECEK', 'TOPLAM']
PAYMENT_KEYWORDS = ['NAKIT', 'KREDI KARTI', 'KREDI', 'KART']
CHANGE_KEYWORDS = ['PARA USTU', 'PARAUSTU', 'PARA USTÜ', 'USTU']


def normalize_market_name(name):
    if not name:
        return None
    n = normalize_tr(name)
    if 'A101' in n or 'A 101' in n:
        return 'A101'
    if 'BIM' in n or 'B1M' in n:
        return 'BİM'
    if 'MIGROS' in n:
        return 'MİGROS'
    if 'SOK' in n:
        return 'ŞOK'
    return name.strip()


def extract_first(patterns, text):
    for pat in patterns:
        m = re.search(pat, text, flags=re.IGNORECASE | re.DOTALL)
        if m:
            return m.group(1).strip()
    return None


def extract_date(normalized_text):
    return extract_first(DATE_PATTERNS, normalized_text)


def extract_time(normalized_text):
    return extract_first(TIME_PATTERNS, normalized_text)


def parse_amount(value_str):
    """Türk para formatını float'a çevirir: 1.250,90 → 1250.90."""
    if not value_str:
        return None

    v = str(value_str).strip()
    v = re.sub(r'[TL₺*\s]', '', v)

    if ',' in v and '.' in v:
        v = v.replace('.', '').replace(',', '.')
    elif ',' in v:
        v = v.replace(',', '.')

    v = re.sub(r'[^0-9.]', '', v)

    if v.count('.') > 1:
        parts = v.split('.')
        v = ''.join(parts[:-1]) + '.' + parts[-1]

    try:
        result = float(v)
        return result if result >= 0 else None
    except ValueError:
        return None


def extract_amounts_for_keyword(normalized_text, keywords, window=100):
    candidates = []
    for kw in keywords:
        kw_pat = re.escape(kw)
        pat = rf'{kw_pat}.{{0,{window}}}?{AMT}'
        for m in re.finditer(pat, normalized_text, flags=re.DOTALL | re.IGNORECASE):
            amt = parse_amount(m.group(1))
            if amt is not None:
                candidates.append({
                    'keyword': kw,
                    'amount': amt,
                    'snippet': re.sub(r'\s+', ' ', m.group(0))[:120],
                })
    return candidates


def select_best_amount(candidates, priority_keywords=None, prefer='priority'):
    if not candidates:
        return None
    if prefer == 'max':
        return max(candidates, key=lambda x: x['amount'])['amount']

    priority_keywords = priority_keywords or []
    for kw in priority_keywords:
        for c in candidates:
            if c.get('keyword') == kw:
                return c['amount']

    return max(candidates, key=lambda x: x['amount'])['amount']



def _make_candidate(keyword, amount, snippet, method='regex', rate=None, priority=50):
    """Standart KDV aday kaydı üretir."""
    if amount is None:
        return None
    return {
        'keyword': keyword,
        'amount': float(amount),
        'snippet': re.sub(r'\s+', ' ', str(snippet))[:140],
        'method': method,
        'rate': rate,
        'priority': priority,
    }


def _amounts_in_line(line):
    """Bir satırdaki para formatına benzeyen tüm değerleri float olarak döndürür."""
    vals = []
    for raw in re.findall(AMT, line):
        amt = parse_amount(raw)
        if amt is not None:
            vals.append(amt)
    return vals


def _rate_from_line(line):
    """KDV oranını satırdan yakalar: %1, %8, %10, %18, %20 vb."""
    m = re.search(r'(?:%|\*)\s*(1|8|10|18|20)\b', line)
    if m:
        return int(m.group(1))
    m = re.search(r'\bKDV\s*(?:ORANI)?\s*(1|8|10|18|20)\b', line)
    if m:
        return int(m.group(1))
    return None


def extract_vat_breakdown(normalized_text):
    """
    KDV çıkarımı — V5.1 güçlendirilmiş sürüm.

    Neden eski sürüm yetersizdi?
    - Sadece %1/%8/%18 oranlarını arıyordu.
    - KDV tablosunda "matrah + KDV tutarı" aynı satırdaysa çoğu zaman yanlış tutarı seçiyordu.
    - Bazı market fişlerinde KDV toplamı doğrudan değil, satır satır veriliyor.

    Bu sürüm:
    - %1/%8/%10/%18/%20 oranlarını destekler.
    - "TOPLAM KDV / KDV TUTARI / TOPKDV" gibi doğrudan satırları yakalar.
    - KDV tablosu satırında birden fazla tutar varsa son tutarı KDV tutarı adayı sayar.
    - KDV adaylarını saklar; nihai seçim toplam tutar bilindikten sonra yapılır.
    """
    vat = {
        'kdv_1': None, 'kdv_8': None, 'kdv_10': None, 'kdv_18': None, 'kdv_20': None,
        'toplam_kdv': None,
        'kdv_adaylari': []
    }

    lines = [ln.strip() for ln in str(normalized_text).split('\n') if str(ln).strip()]

    # 1) Doğrudan toplam KDV satırları.
    direct_keywords = [
        'TOPLAM KDV', 'KDV TOPLAM', 'KDV TUTARI', 'TOPKDV',
        'HESAPLANAN KDV', 'TOP. KDV', 'TOP KDV'
    ]
    direct_candidates = extract_amounts_for_keyword(normalized_text, direct_keywords, window=90)
    for c in direct_candidates:
        c['method'] = 'direct_total_keyword'
        c['priority'] = 10
    vat['kdv_adaylari'].extend(direct_candidates)

    # 2) Oran bazlı açık kalıplar.
    for rate in [1, 8, 10, 18, 20]:
        patterns = [
            rf'%\s*{rate}\s*(?:KDV|KD V)?\s*[:\-]?\s*{AMT}',
            rf'KDV\s*%?\s*{rate}\s*[:\-]?\s*{AMT}',
            rf'KDV\s*(?:ORANI)?\s*{rate}\s*[:\-]?\s*{AMT}',
        ]
        for pat in patterns:
            m = re.search(pat, normalized_text, flags=re.IGNORECASE)
            if m:
                amt = parse_amount(m.group(1))
                vat[f'kdv_{rate}'] = amt
                cand = _make_candidate(
                    keyword=f'KDV %{rate}',
                    amount=amt,
                    snippet=m.group(0),
                    method='rate_keyword_regex',
                    rate=rate,
                    priority=20
                )
                if cand:
                    vat['kdv_adaylari'].append(cand)
                break

    # 3) Satır bazlı tablo fallback.
    # Örnekler:
    #   "%10 KDV MATRAH 100,00 KDV 10,00" → son tutar 10,00
    #   "*20 240,00 40,00"                → son tutar 40,00
    #   "KDV 20 240,00 40,00"             → son tutar 40,00
    for line in lines:
        norm_line = full_normalize(line)
        rate = _rate_from_line(norm_line)
        amounts = _amounts_in_line(norm_line)

        looks_like_vat_row = (
            ('KDV' in norm_line and len(amounts) >= 1) or
            (rate is not None and len(amounts) >= 2)
        )
        if not looks_like_vat_row:
            continue

        # Tablo satırlarında son para değeri çoğunlukla KDV tutarıdır.
        selected = amounts[-1] if amounts else None
        if selected is None:
            continue

        if rate in [1, 8, 10, 18, 20] and vat.get(f'kdv_{rate}') is None:
            vat[f'kdv_{rate}'] = selected

        cand = _make_candidate(
            keyword=f'KDV %{rate}' if rate else 'KDV SATIRI',
            amount=selected,
            snippet=line,
            method='line_table_last_amount',
            rate=rate,
            priority=30 if rate else 40
        )
        if cand:
            vat['kdv_adaylari'].append(cand)

    # 4) Aynı adayların tekrarlarını temizle.
    unique = []
    seen = set()
    for c in vat['kdv_adaylari']:
        key = (round(float(c['amount']), 2), c.get('method'), c.get('rate'), c.get('snippet'))
        if key not in seen:
            seen.add(key)
            unique.append(c)
    vat['kdv_adaylari'] = unique

    # Toplam tutar bilinmeden güvenli nihai seçim yapmayız.
    # İlk aşamada doğrudan toplam KDV adayı varsa onu geçici olarak seç.
    direct_only = [c for c in vat['kdv_adaylari'] if c.get('method') == 'direct_total_keyword']
    if direct_only:
        vat['toplam_kdv'] = sorted(direct_only, key=lambda x: x.get('priority', 99))[0]['amount']
    else:
        parts = [vat.get('kdv_1'), vat.get('kdv_8'), vat.get('kdv_10'), vat.get('kdv_18'), vat.get('kdv_20')]
        if any(x is not None for x in parts):
            vat['toplam_kdv'] = round(sum(x or 0 for x in parts), 2)

    return vat


def select_best_vat_amount(vat_info, general_total=None):
    """
    KDV adayları arasından nihai tutarı seçer.

    Ana mantık:
    - KDV, toplam tutardan büyük/eşit olamaz.
    - Türkiye perakende fişlerinde KDV/toplam oranı çoğunlukla %0–%25 bandındadır.
    - Birden fazla oran kırılımı varsa toplamları tercih edilir.
    - Doğrudan "TOPLAM KDV" adayı varsa güçlü kanıttır.
    """
    candidates = list(vat_info.get('kdv_adaylari', []))

    def valid_amount(x):
        if x is None or x <= 0:
            return False
        if general_total is not None:
            if x >= general_total - 0.02:
                return False
            if x > general_total * 0.30:
                return False
        return True

    # Oran kırılımlarının toplamı çoğu zaman en güvenilir fallback'tir.
    parts = [vat_info.get('kdv_1'), vat_info.get('kdv_8'), vat_info.get('kdv_10'),
             vat_info.get('kdv_18'), vat_info.get('kdv_20')]
    part_values = [x for x in parts if valid_amount(x)]
    if len(part_values) >= 2:
        summed = round(sum(part_values), 2)
        if valid_amount(summed):
            return summed, 'sum_of_rate_breakdown'

    # Geçerli adayları filtrele.
    valid_candidates = [c for c in candidates if valid_amount(c.get('amount'))]
    if not valid_candidates:
        # Tek oran kırılımı varsa onu kullan.
        if len(part_values) == 1:
            return round(part_values[0], 2), 'single_rate_breakdown'
        return None, 'not_found'

    # Öncelik: doğrudan toplam KDV > oran regex > tablo satırı.
    valid_candidates = sorted(
        valid_candidates,
        key=lambda c: (
            c.get('priority', 99),
            0 if c.get('method') == 'direct_total_keyword' else 1,
            c.get('amount', 0)
        )
    )
    return round(float(valid_candidates[0]['amount']), 2), valid_candidates[0].get('method', 'candidate')


def _sanitize_vat_vs_total(vat_value, general_total):
    """
    KDV değeri genel toplama eşit veya toplamın mantıksız büyük oranıysa yanlış eşleşme kabul edilir.
    """
    if vat_value is None:
        return None
    if general_total is None:
        return vat_value
    if abs(vat_value - general_total) < 0.02:
        return None
    if vat_value >= general_total:
        return None
    if vat_value > general_total * 0.30:
        return None
    return vat_value


def extract_payment(normalized_text):
    payment_candidates = extract_amounts_for_keyword(normalized_text, PAYMENT_KEYWORDS, window=80)
    payment_type = None

    if re.search(r'\bNAKIT\b', normalized_text):
        payment_type = 'NAKİT'
    elif re.search(r'\b(KREDI|KART)\b', normalized_text):
        payment_type = 'KREDİ/KART'

    payment_amount = select_best_amount(payment_candidates, PAYMENT_KEYWORDS)
    return payment_type, payment_amount, payment_candidates


def extract_address_from_lines(lines, market_name=None):
    """
    Adres standardı fişten fişe değişir.
    Bu yüzden mağaza üst bloğundaki adres ipuçları aranır.
    """
    addr_keywords = [
        'MAH', 'MAHALLE', 'CAD', 'CADDE', 'SOK', 'SOKAK',
        'NO:', 'NO ', 'APT', 'BLOK', 'ILCE', 'IL ', 'ANKARA',
        'ISTANBUL', 'IZMIR', 'BURSA', 'KONYA', 'ADANA', 'ANTALYA'
    ]

    selected = []
    for line in lines[:14]:
        n = full_normalize(line)
        if any(k in n for k in addr_keywords):
            if not any(k in n for k in ['TOPLAM', 'KDV', 'NAKIT', 'KREDI', 'PARA USTU']):
                selected.append(line.strip())

    return ' | '.join(selected) if selected else None


def is_financial_or_header_line(norm_line):
    """
    Ürün satırı olmayan satırları filtreler.
    Finansal, başlık, birim fiyat ve tarih-saat karışımı satırlar elenir.
    """
    # Kesin finansal / başlık anahtar kelimeleri
    blocked = [
        'TOPLAM', 'KDV', 'NAKIT', 'KREDI', 'KART', 'PARA USTU',
        'TARIH', 'SAAT', 'FIS', 'BELGE', 'VERGI', 'KASIYER',
        'ARA TOPLAM', 'GENEL', 'SUBE', 'MAGAZA', 'TESEKKUR',
        'SEPET', 'BARKOD', 'INDIRIM', 'KAMPANYA',
    ]
    if any(k in norm_line for k in blocked):
        return True

    # Birim fiyat satırları: "X 34,95 TL/AD", "2 AD 17,90 TL/AD"
    # Bu satırlar fiyat birimidir — ürün adı değildir.
    if re.search(r'\bTL\s*/\s*(?:AD|KG|LT|MT|GR|ADET)\b', norm_line):
        return True

    # Saf çarpım satırları: "X 34,95" veya "* 34,95" — adet x fiyat gösterimi
    if re.match(r'^\s*[X\*]\s*\d', norm_line):
        return True

    # Tarih-saat karışımı satırlar: "19/04/26 15.29 AC-04"
    if re.search(r'\b\d{2}/\d{2}/\d{2}\b.*\d{2}[.:]\d{2}', norm_line):
        return True

    return False


def extract_item_lines(lines):
    """
    Ürün satırı parser.

    Yaklaşım:
    - Satırda para deseni olmalı.
    - Finansal/başlık satırları elenir.
    - Ürün adı son para tutarından önceki metinden çıkarılır.
    - Adet bilgisi 2 X / 2 ADET / 2* gibi ipuçlarından yakalanır.
    """
    items = []

    for line in lines:
        original = str(line).strip()
        if not original:
            continue

        n = full_normalize(original)
        if is_financial_or_header_line(n):
            continue

        amounts = re.findall(AMT, n)
        if not amounts:
            continue

        price = parse_amount(amounts[-1])
        if price is None:
            continue

        # Ürün adı: son tutarı ve olası adet/x işaretlerini çıkar.
        product_part = re.sub(AMT + r'\s*$', '', original).strip(' -:*')
        product_part_clean = re.sub(r'\b\d{1,3}\s*(?:X|ADET|AD|\*)\b', '', full_normalize(product_part)).strip()
        product_letters = re.sub(r'[^A-ZÇĞIİÖŞÜ]', '', product_part_clean)

        if len(product_letters) < 2:
            continue

        qty = 1
        q = re.search(r'\b(\d{1,3})\s*(?:X|ADET|AD|\*)\b', n)
        if q:
            qty = int(q.group(1))

        items.append({
            'urun_adi': product_part.strip(),
            'adet': qty,
            'fiyat': price,
            'ham_satir': original,
        })

    return items


def compute_field_confidence(fields, raw_text, ocr_rows, ocr_score):
    """
    Alan bazlı confidence score.
    Bu skor model tahmini değil; rule/evidence tabanlı kalite göstergesidir.
    """
    norm = full_normalize(raw_text)
    used = [r for r in ocr_rows if r.get('used_in_text')]
    avg_conf = _safe_mean([r.get('confidence', 0) for r in used], 0.0)
    base = min(float(ocr_score or 0) / 250, 1.0)

    def score_field(field, keywords=None, value_bonus=0.45):
        value = fields.get(field)
        if value is None or value == [] or value == '':
            return 0.0

        kw_hit = 0.0
        if keywords:
            kw_hit = 0.25 if any(k in norm for k in keywords) else 0.0

        s = value_bonus + kw_hit + avg_conf * 0.20 + base * 0.10
        return round(float(min(s, 1.0)), 3)

    return {
        'market_adi':   score_field('market_adi', MARKET_KEYWORDS),
        'vergi_no':     score_field('vergi_no', ['VERGI', 'VKN', 'VD']),
        'fis_no':       score_field('fis_no', ['FIS', 'FIŞ', 'BELGE', 'NO']),
        'tarih':        score_field('tarih', ['TARIH']),
        'saat':         score_field('saat', ['SAAT']),
        'urunler':      round(min(1.0, 0.20 + len(fields.get('urunler', [])) * 0.15 + avg_conf * 0.25), 3) if fields.get('urunler') else 0.0,
        'ara_toplam':   score_field('ara_toplam', SUBTOTAL_KEYWORDS),
        'kdv':          score_field('kdv', ['KDV']),
        'genel_toplam': score_field('genel_toplam', GENERAL_TOTAL_KEYWORDS),
        'odeme_tipi':   score_field('odeme_tipi', PAYMENT_KEYWORDS, value_bonus=0.35),
        'odeme_tutari': score_field('odeme_tutari', PAYMENT_KEYWORDS),
        'para_ustu':    score_field('para_ustu', CHANGE_KEYWORDS),
        'kasiyer':      score_field('kasiyer', ['KASIYER', 'KASİYER']),
    }


def validate_receipt_fields(fields):
    """
    Multi-field consistency check.
    Amaç: Çıkarılan alanların birbirleriyle mantıklı olup olmadığını görmek.
    """
    warnings_ = []

    subtotal = fields.get('ara_toplam')
    total = fields.get('genel_toplam') or fields.get('toplam_tutar')
    vat = fields.get('toplam_kdv') or fields.get('kdv')
    payment = fields.get('odeme_tutari')
    change = fields.get('para_ustu')

    if subtotal is not None and total is not None and subtotal > total + 1.0:
        warnings_.append('ara_toplam_genel_toplamdan_buyuk')

    if payment is not None and total is not None:
        if payment + 0.05 < total:
            warnings_.append('odeme_tutari_toplamdan_kucuk')
        if change is not None and abs((payment - total) - change) > 1.0:
            warnings_.append('para_ustu_tutari_tutarsiz')

    vat_parts = [fields.get('kdv_1'), fields.get('kdv_8'), fields.get('kdv_18')]
    if vat is not None and any(v is not None for v in vat_parts):
        vat_sum = round(sum(v or 0 for v in vat_parts), 2)
        if abs(vat_sum - vat) > 1.0:
            warnings_.append('kdv_kirilimlari_toplam_kdv_ile_tutarsiz')

    if total is not None and total <= 0:
        warnings_.append('genel_toplam_gecersiz')

    if len(fields.get('urunler', [])) == 0:
        warnings_.append('urun_satiri_bulunamadi')

    return warnings_


def parse_fields(raw_text, ocr_rows=None, ocr_score=0):
    """
    Ham OCR metninden geniş fiş şemasını çıkarır.
    Bulunamayan alanlar None kalır; pipeline asla bu yüzden çökmez.
    """
    if not raw_text or not str(raw_text).strip():
        return {}

    ocr_rows = ocr_rows or []
    norm = full_normalize(raw_text)
    lines = [ln.strip() for ln in str(raw_text).split('\n') if str(ln).strip()]

    market_raw = extract_first(MARKET_PATTERNS, norm)
    market_name = normalize_market_name(market_raw)
    date = extract_date(norm)
    time = extract_time(norm)

    subtotal_candidates = extract_amounts_for_keyword(norm, SUBTOTAL_KEYWORDS, window=80)
    general_total_candidates = extract_amounts_for_keyword(norm, GENERAL_TOTAL_KEYWORDS, window=100)
    change_candidates = extract_amounts_for_keyword(norm, CHANGE_KEYWORDS, window=80)

    vat_info = extract_vat_breakdown(norm)
    payment_type, payment_amount, payment_candidates = extract_payment(norm)

    receipt_no = extract_first(RECEIPT_NO_PATTERNS, norm)
    tax_no = extract_first(TAX_PATTERNS, norm)
    cashier = extract_first(CASHIER_PATTERNS, norm)

    subtotal = select_best_amount(subtotal_candidates, SUBTOTAL_KEYWORDS)
    general_total = select_best_amount(general_total_candidates, GENERAL_TOTAL_KEYWORDS, prefer='max')
    change_amount = select_best_amount(change_candidates, CHANGE_KEYWORDS)

    items = extract_item_lines(lines)
    address = extract_address_from_lines(lines, market_name=market_name)

    selected_vat, selected_vat_method = select_best_vat_amount(vat_info, general_total=general_total)

    fields = {
        'market_adi': market_name,
        'adres': address,
        'vergi_no': tax_no,
        'fis_no': receipt_no,
        'tarih': date,
        'saat': time,
        'tarih_saat': f'{date} {time}' if date and time else date,
        'urunler': items,
        'urun_sayisi': len(items),
        'ara_toplam': subtotal,
        'kdv_1': vat_info.get('kdv_1'),
        'kdv_8': vat_info.get('kdv_8'),
        'kdv_10': vat_info.get('kdv_10'),
        'kdv_18': vat_info.get('kdv_18'),
        'kdv_20': vat_info.get('kdv_20'),
        'toplam_kdv': _sanitize_vat_vs_total(selected_vat, general_total),
        'kdv': _sanitize_vat_vs_total(selected_vat, general_total),
        'kdv_secim_yontemi': selected_vat_method,
        'genel_toplam': general_total,
        'toplam_tutar': general_total,
        'odeme_tipi': payment_type,
        'odeme_tutari': payment_amount,
        'para_ustu': change_amount,
        'kasiyer': cashier,
        'toplam_adaylari': general_total_candidates,
        'kdv_adaylari': vat_info['kdv_adaylari'],
        'odeme_adaylari': payment_candidates,
    }

    fields['field_confidence'] = compute_field_confidence(fields, raw_text, ocr_rows, ocr_score)
    fields['validation_warnings'] = validate_receipt_fields(fields)
    fields['validation_status'] = 'pass' if not fields['validation_warnings'] else 'review'

    return fields


if image_paths:
    sample_fields = parse_fields(sample_text, sample_rows, chosen_score)
    print('V4 Parsing Sonuçları — Geniş Şema + Confidence + Validation')
    for key in ['market_adi', 'adres', 'vergi_no', 'fis_no', 'tarih_saat', 'urun_sayisi',
                'ara_toplam', 'kdv_1', 'kdv_8', 'kdv_10', 'kdv_18', 'kdv_20', 'toplam_kdv', 'kdv_secim_yontemi',
                'genel_toplam', 'odeme_tipi', 'odeme_tutari', 'para_ustu', 'kasiyer',
                'validation_status', 'validation_warnings']:
        print(f'  {key:<22}: {sample_fields.get(key)}')

    print('\nAlan Bazlı Confidence:')
    display(pd.DataFrame([sample_fields.get('field_confidence', {})]).T.rename(columns={0: 'confidence'}))

    print('\nÜrün örnekleri:')
    display(pd.DataFrame(sample_fields.get('urunler', [])).head())
````

- eksik alanları

tek bir kayıt altında saklar.


````python

def process_single_receipt(image_path):
    """
    Tek fiş için uçtan uca V4 pipeline.
    """
    image_path = Path(image_path)
    filename = image_path.name

    record = {
        'dosya_adi': filename,
        'durum': 'basarili',
        'hata_mesaji': None,
        'market_adi': None,
        'adres': None,
        'vergi_no': None,
        'fis_no': None,
        'tarih': None,
        'saat': None,
        'tarih_saat': None,
        'urunler': [],
        'urun_sayisi': 0,
        'ara_toplam': None,
        'kdv_1': None,
        'kdv_8': None,
        'kdv_10': None,
        'kdv_18': None,
        'kdv_20': None,
        'toplam_kdv': None,
        'kdv_secim_yontemi': None,
        'kdv': None,
        'genel_toplam': None,
        'toplam_tutar': None,
        'odeme_tipi': None,
        'odeme_tutari': None,
        'para_ustu': None,
        'kasiyer': None,
        'eksik_alanlar': [],
        'ocr_satir_sayisi': 0,
        'dusuk_guven_sayisi': 0,
        'secilen_strateji': None,
        'ocr_kalite_skoru': None,
        'ham_ocr_metni': None,
        'toplam_adaylari': [],
        'kdv_adaylari': [],
        'odeme_adaylari': [],
        'ocr_aday_skorlari': [],
        'field_confidence': {},
        'validation_status': None,
        'validation_warnings': [],
    }

    # ADIM 1: Görüntü yükleme + 16 aday üretimi
    try:
        bgr, _, strategies = preprocess_all_strategies(image_path)
        if bgr is None:
            raise ValueError('Görsel okunamadı veya boş.')
    except Exception as e:
        record['durum'] = 'hata'
        record['hata_mesaji'] = f'preprocessing_hatasi: {e}'
        record['eksik_alanlar'] = ['tarih', 'toplam_tutar', 'kdv']
        return record, []

    # ADIM 2: Akıllı OCR best-of-candidates
    try:
        raw_text, ocr_rows, chosen_strategy, score, candidate_scores = run_ocr_best_strategy(
            bgr, strategies, return_all_candidates=True
        )

        record['ham_ocr_metni'] = raw_text
        record['ocr_satir_sayisi'] = len(ocr_rows)
        record['dusuk_guven_sayisi'] = sum(1 for r in ocr_rows if r.get('low_confidence'))
        record['secilen_strateji'] = chosen_strategy
        record['ocr_kalite_skoru'] = score
        record['ocr_aday_skorlari'] = candidate_scores

        if not raw_text.strip():
            record['durum'] = 'uyari'
            record['hata_mesaji'] = 'ocr_bos: OCR boş metin döndürdü'
            record['eksik_alanlar'] = ['tarih', 'toplam_tutar', 'kdv']
            return record, ocr_rows

    except Exception as e:
        record['durum'] = 'hata'
        record['hata_mesaji'] = f'ocr_hatasi: {e}'
        record['eksik_alanlar'] = ['tarih', 'toplam_tutar', 'kdv']
        return record, []

    # ADIM 3: Regex + heuristic parsing + confidence + validation
    try:
        fields = parse_fields(raw_text, ocr_rows=ocr_rows, ocr_score=score)
        for key, value in fields.items():
            if key in record:
                record[key] = value
    except Exception as e:
        record['durum'] = 'kismi_basarili'
        record['hata_mesaji'] = f'regex_hatasi: {e}'

    # ADIM 4: Eksik temel alan tespiti — görev kriterleri açısından kritik üçlü
    required_core = ['tarih', 'toplam_tutar', 'kdv']
    eksik = [field for field in required_core if record.get(field) is None]
    record['eksik_alanlar'] = eksik

    if eksik and record['durum'] == 'basarili':
        record['durum'] = 'kismi_basarili'

    # Validation warning varsa hata değil, review sinyali olarak saklanır.
    if record.get('validation_status') == 'review' and record['durum'] == 'basarili':
        record['durum'] = 'basarili_review'

    # OCR satırlarına dosya ve seçilen strateji bilgisini ekle
    for row in ocr_rows:
        row['dosya_adi'] = filename
        row['secilen_strateji'] = chosen_strategy
        row['ocr_kalite_skoru'] = score

    return record, ocr_rows


if image_paths:
    sample_record, sample_rows_out = process_single_receipt(image_paths[0])
    print('\nTek Fiş Pipeline Sonucu — V4')
    for k, v in sample_record.items():
        if k not in ['ham_ocr_metni', 'urunler', 'toplam_adaylari', 'kdv_adaylari',
                     'odeme_adaylari', 'ocr_aday_skorlari']:
            print(f'  {k:<25}: {v}')

    print('\nÜrün satırı örnekleri:')
    display(pd.DataFrame(sample_record.get('urunler', [])).head())

    print('\nOCR aday skorları — ilk 8:')
    display(pd.DataFrame(sample_record.get('ocr_aday_skorlari', [])).sort_values('score', ascending=False).head(8))

    print(f'\nHam OCR metni — ilk 600 karakter:')
    print(str(sample_record['ham_ocr_metni'])[:600])
````

## 17. Hata Analizi

Hata kaynakları kategorize edilir. Bu analiz, hangi adımda iyileştirme yapılması gerektiğini gösterir.

````python
ERROR_LABELS = {
    'preprocessing_hatasi': 'Görüntü ön işleme hatası — dosya okunamadı',
    'ocr_hatasi':           'OCR motor hatası — EasyOCR çöktü',
    'ocr_bos':              'OCR boş metin döndürdü — görsel tanınamadı',
    'regex_hatasi':         'Regex/parsing hatası',
}

def classify_error(msg):
    if not msg:
        return 'eksik_alan'
    msg = str(msg).lower()
    for key in ERROR_LABELS:
        if key in msg:
            return key
    return 'diger'

problem_rows = df_results[df_results['durum'].isin(['hata', 'uyari', 'kismi_basarili'])].copy()

if problem_rows.empty:
    print('✓ Tüm fişler hatasız işlendi!')
else:
    problem_rows['hata_kategorisi'] = problem_rows['hata_mesaji'].apply(classify_error)
    # Eksik alan problemi
    eksik_mask = df_results['eksik_alanlar'].apply(lambda x: isinstance(x, list) and len(x) > 0)
    df_eksik = df_results[eksik_mask][['dosya_adi', 'durum', 'eksik_alanlar']]

    print(f'Sorunlu fiş sayısı: {len(problem_rows)} / {len(df_results)}')
    print('\n[Hata Kategorileri]')
    display(problem_rows[['dosya_adi', 'durum', 'hata_mesaji', 'hata_kategorisi']].head(10))

    print('\n[Eksik Alan İçeren Fişler]')
    display(df_eksik.head(10))
````

## 18. Değerlendirme — Teknik Özet

---

### Bu Sistem Ne Yapıyor?

Ham market fişi fotoğraflarından yapılandırılmış fiş bilgisi çıkarır:

```text
MARKET ADI → ADRES → VERGİ NO → FİŞ NO → TARİH/SAAT → ÜRÜNLER → KDV → TOPLAM → ÖDEME
```

Görevde istenen temel alanlar korunmuştur:

- Tarih
- Toplam Tutar
- KDV
- DataFrame / JSON output

Ek olarak ürün satırları, ödeme bilgisi, kasiyer ve OCR kalite sinyalleri de çıkarılır.

---

### Tanılanan Ana OCR Sorunu

Bazı fiş fotoğraflarında OCR çıktısı anlamlı kelimeler yerine tek karakterler döndürebilir:

```text
8, ğ, {, ;, | gibi kısa ve anlamsız karakterler
```

Bu genellikle iki nedenden kaynaklanır:

1. Fotoğrafın 90/180/270 derece dönmüş olması
2. Preprocessing'in metin çizgilerini bozması

---

### Uygulanan Çözüm

Bu yüzden notebook tek bir preprocessing çıktısına bağlı kalmaz:

```text
4 rotasyon × 4 preprocessing = 16 OCR adayı
```

Her aday OCR'a verilir ve şu sinyallerle skorlanır:

- anlamlı fiş kelimesi sayısı
- para formatı sayısı
- tarih/saat formatı sayısı
- OCR confidence
- tek karakter gürültü oranı

En yüksek skorlu aday final metin kabul edilir.

---

### Neden Regex Fallback Sistemi?

Market fişleri standart değildir. Aynı kavram farklı şekillerde yazılabilir:

```text
TOPLAM, GENEL TOPLAM, TOPLAM TUTAR, ODENECEK
KDV, K.D.V, TOPKDV, TOPLAM KDV
FIS NO, BELGE NO, NO
```

Bu nedenle tek regex yerine çoklu fallback pattern kullanılmıştır.

---

### Neden Ürün Satırı Parsing Eklendi?

Görev temel olarak Tarih/Toplam/KDV ister; ancak profesyonel bir fiş okuma sisteminde ürün satırlarının da listelenmesi beklenir. Bu notebook ürün adı, adet ve fiyat için heuristik bir ürün satırı çıkarımı yapar.

---

### Limitasyonlar

- Çok bulanık, kesilmiş veya aşırı düşük çözünürlüklü fişlerde OCR başarısı düşebilir.
- Adres ve kasiyer alanları market şablonlarına göre değiştiği için her zaman bulunamayabilir.
- Ürün satırı parsing regex/heuristic tabanlıdır; karmaşık kampanya satırlarında hatalı eşleşme olabilir.
